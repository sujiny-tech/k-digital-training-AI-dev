### Pandas ÌôúÏö© 5. groupby ‚ú®

+ Split : ÌäπÏ†ïÌïú Í∏∞Ï§ÄÏùÑ Î∞îÌÉïÏúºÎ°ú DataFrameÏùÑ Î∂ÑÌï†
+ Apply : ÌÜµÍ≥ÑÌï®Ïàò (sum(), mean(), median()..)ÏùÑ Ï†ÅÏö©Ìï¥ÏÑú Í∞Å Îç∞Ïù¥ÌÑ∞ ÏïïÏ∂ï
+ Combine : ApplyÎêú Í≤∞Í≥ºÎ•º Î∞îÌÉïÏúºÎ°ú ÏÉàÎ°úÏö¥ SeriesÎ•º ÏÉùÏÑ±(group_key:applied_value)

```python
covid
```

```python
#WHO RegionÎ≥Ñ ÌôïÏßÑÏûêÏàò

# 1. covid ÏóêÏÑú ÌôïÏßÑÏûê Ïàò columnÎßå Ï∂îÏ∂ú
# 2. Ïù¥Î•º covidÏùò WHO RegionÏùÑ Í∏∞Ï§ÄÏúºÎ°ú groupby Ïã§Ìñâ

covid_by_region=covid['Confirmed'].groupby(by=covid["WHO Region"])

covid_by_region
```

```python
covid_by_region.sum()
```

```python
# Íµ≠Í∞ÄÎãπ Í∞êÏóºÏûê Ïàò

covid_by_region.mean() #sum()/Íµ≠Í∞ÄÏàò
```
